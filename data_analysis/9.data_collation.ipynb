## 更多合并DataFrame例子

```python
customers_data = {
    '客户ID': [1, 2, 3, 4, 5],
    '姓名': ['Tom', 'Bob', 'Mary', 'Alice', 'John'],
    '年龄': [20, 35, 27, 19, 42]
}
customer_df4 = pd.DataFrame(customers_data)


orders_data = {
    '订单ID': [1001, 1002, 1003, 1004, 1005, 1006],
    '订单日期': ['2000-03-12', '2000-03-13', '2000-03-13', '2000-03-15', '2000-03-18', '2000-03-21'],
    '客户ID': [1, 1, 2, 6, 5, 3],
    '产品': ['A', 'B', 'C', 'D', 'E', 'F'],
    '数量': [2, 3, 1, 4, 5, 2]
}
order_df4 = pd.DataFrame(orders_data)
```

```python
pd.merge(customer_df4, order_df4, on='客户ID', how='left')
```

### right合并customer_df4和order_df4

# 对DataFrame进行分组运算

```python
df = pd.DataFrame({
    '分店编号': ['001', '002', '001', '002', '001', '002', '001', '002'],
    '时间段': ['2022Q1', '2022Q1', '2022Q1', '2022Q1', '2022Q2', '2022Q2', '2022Q2', '2022Q2'],
    '商品类别': ['生鲜食品', '生鲜食品', '休闲食品', '休闲食品', '生鲜食品', '生鲜食品', '休闲食品', '休闲食品'],
    '销售额': [1500, 2000, 3000, 2500, 1800, 2200, 3200, 2700],
    '销售数量': [105,  84, 171, 162,  67, 150,  99,  57]
})
```

```python
df
```

## 把df针对分店编号进行分组，计算销售额平均值

```python
df.groupby("分店编号")["销售额"].mean()
```

## 把df针对分店编号进行分组，计算销售额和销售数量的平均值

```python
df.groupby("分店编号")[["销售额", "销售数量"]].mean()
```

## 把df针对分店编号和时间段进行分组，计算销售额和销售数量的平均值

```python
df.groupby(["分店编号", "时间段"])[["销售额", "销售数量"]].mean()
```

```python
def max_plus_10(nums):
    return nums.max() + 10
```

## 把df针对分店编号和时间段进行分组，对销售额和销售数量进行max_plus_10操作

```python
df.groupby(["分店编号", "时间段"])[["销售额", "销售数量"]].apply(max_plus_10)
```

## 把df的商品类别作为索引，分店编号作为列，计算销售额的平均值

```python
pd.pivot_table(df, index='商品类别', columns='分店编号', values='销售额', aggfunc=np.mean)
```

## 把df的商品类别作为索引，分店编号作为列，计算销售额的总和

```python
pd.pivot_table(df, index='商品类别', columns='分店编号', values='销售额', aggfunc=np.sum)
```

```python
pd.pivot_table(df, index='商品类别', columns='分店编号', values='销售额', aggfunc="sum")
```

*注意：以上红色警告意思是当前使用的 Pandas 版本中，aggfunc=np.sum 和 aggfunc="sum" 的计算结果完全相同。但是未来pandas版本你需要更改写法以实现sum的作用，只需将 np.sum 改为 "sum"<br>这与之前学习的aplymap与map警告一样pandas会逐步淘汰以前的旧写法虽说现在的版本依然能使用，Pandas 正在逐步引导开发者放弃一些旧的、模糊的、性能不佳的写法，转向更明确、更高效、更符合“Pandas 风格”的新写法*

# 进行数据分箱

```python
df1 = pd.read_csv(r"F:\BaiduNetdiskDownload\数据分析课程操作的文件\05数据整理篇\residents_data.csv")
df1
```

```python
df1.groupby("年龄")["工资"].mean()
```

```python
age_bins = [0, 10, 20, 30, 40, 50, 60, 120]
pd.cut(df1["年龄"], age_bins)
```

```python
age_labels = ['儿童', '青少年', '青年', '壮年', '中年', '中老年', '老年'] # 定义分组标签
df1["年龄组"] = pd.cut(df1["年龄"], age_bins, labels=age_labels)
df1
```

```python
df1.groupby("年龄组")["工资"].mean()
```

**这个警告是因为 Pandas 未来会改变 groupby 对分类数据的默认处理方式，建议你现在就明确指定 observed=True。**

```python
df1.groupby("年龄组", observed=True)["工资"].mean()
```
